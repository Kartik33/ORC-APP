```python
# %load read_xml.py
#%matplotlib notebook
import xml.etree.ElementTree as ET
from glob import glob
import cv2
import numpy as np
from matplotlib import pyplot as plt

#this is the path to cordinates which have not been edited 
tree_path = '../resources/Bill2/win_xml/IMG_20171112_190347.xml'
#path to cordinates which have been edited 
edited_tree_path = '../resources/Bill2/wintech_xml/IMG_20171112_190347.xml'

def get_cordinate(path):
    tree = ET.parse(path)
    root = tree.getroot()
    all_cordinates = []
    size = root.find('size')
    img_size = []
    for s in size:
        if s.tag == 'width':
            img_size.append(s.text)
        if s.tag == 'height':
            img_size.append(s.text)
            
    for obj in root.findall('object'):
        #print(obj.find('name').text)
        for bndbox in obj.findall('bndbox'):
            one_box_cordinates = []
            for b in bndbox:
                one_box_cordinates.append(b.text)
            all_cordinates.append(one_box_cordinates)
    return all_cordinates , img_size

def get_xml_list(edited_tree_path , mask_out_path , line):
    all_edited_tree = []
    for imgpath in glob(edited_tree_path+'*.xml'):
        xml_name = imgpath.split('/')[-1]
        name = xml_name.split('.')[0]
        etp = edited_tree_path+xml_name
        edited_tree , img_size = get_cordinate(etp)
        mask_out = mask_out_path+name+'.jpg'
        gen_mask(edited_tree , img_size , line , mask_out)
        print('processed image is '  , xml_name)

        
def get_cordinates_list(edited_tree_path , mask_out_path , line):    
    #print('path is :' , edited_tree_path)
    get_xml_list(edited_tree_path , mask_out_path , line)

#call this mehotd to load the list of all cordinates in memory in format 
#[ [[xmin,ymin,xmax,ymax ] , [xmin,ymin,xmax,ymax]....]  , [  [xmin,ymin,xmax,ymax ] , [xmin,ymin,xmax,ymax]... ] 


#genrate mask here 
def gen_mask(edited_tree , img_size , line ,mask_out):
    img = np.zeros((int(img_size[1]) , int(img_size[0])) , np.uint8)
    print(img.shape)
    for cord in edited_tree:
        x = int(cord[0])
        y = int(cord[1])
        x1 = int(cord[2]) 
        y1 = int(cord[3])
        if line:
            h = y1 - y
            w = x1 - x
            if h > 50:
                a = int((h*0.2)/2)
                b = int(w*0.95)
                b = int((w-b)/2)
                img[y+a:y1-a , x :x1] = 255
                #cv2.rectangle(img,(x,y),(x1,y1),(255,255,255),2)
            else:
                img[y :y1 , x : x1] = 255
        else:
            cv2.rectangle(img,(x,y),(x1,y1),(255,255,255),2)
    print(mask_out)
    cv2.imwrite(mask_out , img)
    #cv2.imwrite('x.jpg' , img)
    
```
