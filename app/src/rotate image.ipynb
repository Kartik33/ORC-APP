```python
import tkinter as tk
from PIL import ImageTk,Image
import imutils
import cv2
from matplotlib import pyplot as plt
from glob import glob
```

```python
import numpy as np
class RotateImage:
    def __init__(self, path):
        self.path = path
        self.currentImgPath = None
        self.counter = 0
        self.panelA = None
        self.img = None
        self.final_angle = 0 
    
    def show_img( self , img , shape):
        root.geometry(str(shape[0])+"x"+str(shape[1]))
        
        img = imutils.resize(img , width=1080 , height=1080)
        
        h,w,ch=img.shape
        spacing=30; linewidth=1
        for hchange in np.arange(0,h,spacing):
            img = cv2.line(img, (0, int(hchange)), (w, int(hchange)), (0,0,255),linewidth)
        for wchange in np.arange(0,w,spacing):
            img = cv2.line(img, (int(wchange),0), (int(wchange),h), (0,0,255),linewidth)
            
        img = self.conv_to_pil(img)
        img = ImageTk.PhotoImage(img)
        panelA = self.panelA
        if panelA is None :
            panelA = tk.Label(image=img)
            panelA.image =  img
            panelA.pack(side="left", padx=10, pady=10)
        else :
            panelA.configure(image=img)
            panelA.image = img
        self.panelA = panelA
    
    
    def open_img(self,path):
        img1 = cv2.imread(path)
        shape = img1.shape
        return imutils.resize(img1 , width=1080 , height=1080) , shape
    
    def conv_to_pil(self , opencv_img):
        image = cv2.cvtColor(opencv_img, cv2.COLOR_BGR2RGB)
        image = Image.fromarray(image)
        return image
    
    def rotate(self,img, angl , shape):
        img = imutils.rotate(image=img , angle=angl)
        self.set_img(img)
        self.final_angle = self.final_angle+angl
        
        self.show_img( self.get_img() , shape)
    
            
    def get_path_list(self , path):
        path_list = []
        for imgpath in glob(path+'*.jpg')+glob(path+'*.png'):
            path_list.append(imgpath)
        return path_list
    
    def save(self ,path):
        print('save angle ' , self.final_angle , path)
        if self.final_angle != 0 :
            
            img = cv2.imread(path)
            img = imutils.rotate(img , angle = self.final_angle)
            x = cv2.imwrite(path , img)
            
            
            print(x , path)
        else:
            print('no change')
        return None
    
    def nxt_path(self , path_list):
        path = path_list[self.counter]
        self.counter = self.counter + 1
        self.set_current_path(path)
        return path
        
    def nxt_img(self,path_list):
        self.save(self.get_current_path())
        self.final_angle = 0
        path = self.nxt_path(path_list)
        img , shape = self.open_img(path)
        self.set_img(img)
        self.show_img(img , shape)
        
    def get_img(self):
        return self.img
    
    def set_img(self,img):
        self.img = img
        
    def set_current_path(self,path):
        self.current_path = path
        
    def get_current_path(self):
        return self.current_path
```

```python
# path = '../resources/anil chem img/box/'
path = '/home/kartik/rasidrev/resources/anil chem img/phone/backup/'

ri = RotateImage(path)
path_list = ri.get_path_list(path)
print(len(path_list))

ri.set_current_path(ri.nxt_path(path_list))

img , shape = ri.open_img(ri.get_current_path())

ri.set_img(img)

ri.panelA = None

root = tk.Tk()

ri.show_img(ri.get_img(),shape)

btn = tk.Button(root, text="right",
                command=lambda : ri.rotate(ri.get_img(), -0.1 , shape))
btn.pack(side="right", expand="yes", padx="10", pady="10")

btn = tk.Button(root, text="left",
                command=lambda: ri.rotate(ri.get_img(),0.1 , shape))
btn.pack(side="right", expand="yes", padx="10", pady="10")

btn = tk.Button(root, text="save",
                command=lambda: ri.save(ri.get_current_path()))
btn.pack(side="right", expand="yes", padx="10", pady="10")

btn = tk.Button(root, text="next",
                command=lambda: ri.nxt_img(path_list) )
btn.pack(side="right", expand="yes", padx="10", pady="10")

root.mainloop()
```
